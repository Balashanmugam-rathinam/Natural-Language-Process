```python
import nltk
from nltk.tokenize import TreebankWordTokenizer
from nltk.corpus import stopwords
from nltk.probability import FreqDist
import matplotlib.pyplot as plt

# Download stopwords set (if not already downloaded)
nltk.download('stopwords')

# Initialize tokenizer and stopwords list
tokenizer = TreebankWordTokenizer()
stop_words = set(stopwords.words('english'))

# Sample text
text = "Natural language processing with Python is interesting and useful. Python makes NLP tasks easier and more efficient."

# Tokenize the text
tokens = tokenizer.tokenize(text)

# Remove stopwords
filtered_tokens = [word for word in tokens if word.lower() not in stop_words and word.isalpha()]

# Print filtered tokens
print("Filtered Tokens (without stopwords):", filtered_tokens)

# Frequency Distribution
fdist = FreqDist(filtered_tokens)

# Print frequency distribution
print("\nWord Frequency Distribution (without stopwords):")

for word, freq in fdist.items():
 print(f"{word}: {freq}")
    
# Plot frequency distribution
fdist.plot(title='Word Frequency Distribution (Without Stopwords)')
plt.show()
```

📄 GitHub Description
This is a simple Natural Language Processing (NLP) project created to help beginners understand how to tokenize text, remove stopwords, and visualize word frequencies using Python. It uses NLTK for tokenization and stopword filtering, and Matplotlib for visualizing the word frequency distribution.

✅ What This Project Does
Uses TreebankWordTokenizer to split text into tokens (words).

Removes common English stopwords using NLTK.

Filters out punctuation and non-alphabetic tokens.

Counts the frequency of each meaningful word.

Displays the frequency data in both text and chart form.
